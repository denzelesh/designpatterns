```python
# MVC_Template_01
# 2014 May 23  by Steven Lipton http://makeAppPie.com
# Controller initializing MVC -- simplest version possible.
# Additional annotations by R Mather - Oct 2018

# MVC EXTENDED WITH DELETE BUTTON

# Tkinter is a Python binding to the Tk GUI toolkit/library - RM
from tkinter import *


# A Model-View-Controller framework for TKinter.
# Model: Data Structure. Controller can send messages to it, and model can respond to message.
# View : User interface elements. Controller can send messages to it. View can call methods from Controller when an event happens.
# Controller: Ties View and Model together. turns UI responses into chages in data.


# Controller: Ties View and Model together.
#       --Performs actions based on View events.
#       --Sends messages to Model and View and gets responses
#       --Has Delegates - RM NOTE: essential for communications between view and model via control

class MyController():
    def __init__(self, parent):
        self.parent = parent
        # CRITICAL PART OF THE MVC
        # RM NOTES:
        # MyController passes a reference to itself (self) when it creates MyModel and MyView objects
        # This allows the MyModel and MyView objects to create 'delegates' (see 'vc' = virtual controller?)
        # The vc objects can then call methods on the MyControl object, thereby allowing the View to update the Model via the Controller
        self.model = MyModel(self)  # initializes the model
        self.view = MyView(self)  # initializes the view
        # Initialize objects in view
        self.view.setEntry_text('Add to Label')  # a non cheat way to do MVC with tkinter control variables
        self.view.setLabel_text('Ready')

    # event handlers
    def quitButtonPressed(self):
        self.parent.destroy()

    def addButtonPressed(self):
        self.view.setLabel_text(self.view.entry_text.get())
        # CRITICAL PART OF THE MVC
        # RM NOTES - In one operation:
        # [1] Get the text from the VIEW;
        # [2] Add it to the MODEL by appending it to the list
        self.model.addToList(self.view.entry_text.get())

    def listChangedDelegate(self):
        # model internally changes and needs to signal a change
        print(self.model.getList())

    ###### **** I ADDED FUNCTION **** #######################################################
    ## It will retrive the item view and get the model to remove it
    def deleteTextButtonPressed(self):
        self.model.removeItemFromList(self.view.entry_text.get())
    ####################################################################################################


# View : User interface elements.
#       --Controller can send messages to it.
#       --View can call methods from Controller vc when an event happens.
#       --NEVER communicates with Model.
#       --Has setters and getters to communicate with controller

class MyView(Frame):

    def __init__(self, vc):
        self.frame = Frame()
        self.frame.grid(row=0, column=0)
        self.vc = vc
        self.entry_text = StringVar()
        self.entry_text.set('nil')
        self.label_text = StringVar()
        self.label_text.set('nil')
        self.loadView()

    def loadView(self):
        quitButton = Button(self.frame, text='Quit', command=self.vc.quitButtonPressed).grid(row=0, column=0)
        addButton = Button(self.frame, text="Add", command=self.vc.addButtonPressed).grid(row=0, column=1)
        entry = Entry(self.frame, textvariable=self.entry_text).grid(row=1, column=0, columnspan=3, sticky=EW)
        label = Label(self.frame, textvariable=self.label_text).grid(row=2, column=0, columnspan=3, sticky=EW)

        ###### **** I ADDED A DELETE BUTTON  **** #######################################################
        deleteTextButton = Button(self.frame, text="Delete", command=self.vc.deleteTextButtonPressed).grid(row=0,
                                                                                                           column=2)
        ##################################################################################################

    def getEntry_text(self):
        # returns a string of the entry text
        return self.entry_text.get()

    def setEntry_text(self, text):
        # sets the entry text given a string
        self.entry_text.set(text)

    def getLabel_text(self):
        # returns a string of the Label text
        return self.label_text.get()

    def setLabel_text(self, text):
        # sets the label text given a string
        self.label_text.set(text)


# Model: Data Structure.
#   --Controller can send messages to it, and model can respond to message.
#   --Uses delegates from vc to send messages to the Control of internal change
#   --NEVER communicates with View
#   --Has setters and getters to communicate with Controller

class MyModel():
    def __init__(self, vc):
        self.vc = vc
        self.myList = ['duck', 'duck', 'goose']
        self.count = 0

    # Delegates-- Model would call this on internal change
    def listChanged(self):
        self.vc.listChangedDelegate()

    # setters and getters
    def getList(self):
        return self.myList

    def initListWithList(self, aList):
        self.myList = aList

    def addToList(self, item):
        print("returned")
        myList = self.myList
        myList.append(item)
        self.myList = myList
        self.listChanged()

    ###### **** I ADDED FUNCTION **** #######################################################
    ## This will remove an item from list
    def removeItemFromList(self, item):
        if item in self.myList:
            self.myList.remove(item)
            self.listChanged()
    #########################################################################################


def main():
    # Create a root window from the Tk GUI class
    root = Tk()
    frame = Frame(root, bg='#0555ff')
    root.title('Hello Penguins')
    app = MyController(root)
    root.mainloop()


if __name__ == '__main__':
    main()

  
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Nicely abstracted structure by which any model can notify any observer (view) of changes in the model
class Subject(object):  # Represents what is being 'observed'

    def __init__(self):
        self._observers = []  # This where references to all the observers are being kept
        # Note that this is a one-to-many relationship: there will be one subject to be observed by multiple _observers

    def attach(self, observer):
        if observer not in self._observers:  # If the observer is not already in the observers list
            self._observers.append(observer)  # append the observer to the list

    def detach(self, observer):  # Simply remove the observer
        try:
            self._observers.remove(observer)
        except ValueError:
            pass

    def notify(self, modifier=None):
        for observer in self._observers:  # For all the observers in the list
            if modifier != observer:  # Don't notify the observer who is actually doing the updating
                observer.update(self)  # Alert the observers!


# Represents the 'data' for which changes will produce notifications to any registered view/observer objects
class Model(Subject):  # Extends the Subject class

    def __init__(self, name):
        Subject.__init__(self)
        self._name = name  # Set the name of the model
        self._labels = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
        self._data = [10, 8, 6, 4, 2, 1]

    @property  # Getter that gets the labels
    def labels(self):
        return self._labels

    @labels.setter  # Setter that sets the labels
    def labels(self, labels):
        self._labels = labels
        self.notify()  # Notify the observers whenever somebody changes the labels

    @property  # Getter that gets the data
    def data(self):
        return self._data

    @data.setter  # Setter that sets the labels
    def data(self, data):
        self._data = data
        self.notify()  # Notify the observers whenever somebody changes the data


# This is the 'standard' view/observer which also acts as an 'abstract' class whereby deriving Bar/Chart/Table views override the update() method
# This 'abstracted' layer is always shown in examples but is important to demonstrate potential polymorphic behaviour of update()
class View():

    def __init__(self, name=""):
        self._name = name  # Set the name of the Viewer

    def update(self, subject):  # Alert method that is invoked when the notify() method in a concrete subject is invoked
        print("Generalised Viewer '{}' has: \nName = {}; \nLabels = {}; \nData = {}".format(self._name, subject._name,
                                                                                            subject._labels,
                                                                                            subject._data))


# Table 'chart' viewer/ConcreteObserver - overrides the update() method
class TableView(View):

    def update(self, subject):  # Alert method that is invoked when the notify() method in a concrete subject is invoked
        fig = plt.figure(dpi=80)
        ax = fig.add_subplot(1, 1, 1)
        table_data = list(map(list, zip(subject._labels, subject._data)))
        table = ax.table(cellText=table_data, loc='center')
        table.set_fontsize(14)
        table.scale(1, 4)
        ax.axis('off')
        plt.show()


# Bar chart viewer/ConcreteObserver - overrides the update() method
class BarView(View):

    def update(self, subject):  # Alert method that is invoked when the notify() method in a concrete subject is invoked
        objects = subject._labels
        y_pos = np.arange(len(objects))
        performance = subject._data
        plt.bar(y_pos, performance, align='center', alpha=0.5)
        plt.xticks(y_pos, objects)
        plt.ylabel('Usage')
        plt.title('Programming language usage')
        plt.show()


######CREATED NEW CLASS####################################################################
#####Creates a new piechart viewer/ConcreteObserver that will overide the update method####
class PieChartView(View):

    def update(self, subject): 
        plt.figure(dpi=80)
        plt.pie(subject._data, labels=subject._labels, autopct='%1.1f%%')
        plt.title('Programming language usage as a pie chart')
        plt.axis('equal')  # so the pie chart is drawn as a circle
        plt.show()
###########################################################################################

# Let's create our subjects
m1 = Model("Model 1")
m2 = Model("Model 2")  # This is never used!

# Let's create our observers
v1 = View("1: standard text viewer")
v2 = TableView("2: table viewer")
v3 = BarView("3: bar chart viewer")
v4 = PieChartView("4: pie chart viewer") ######## Created new observer for piechart
####

# Let's attach our observers to the first model
m1.attach(v1)
m1.attach(v2)
m1.attach(v3)
m1.attach(v4) ######### Ataches the new observer to the model 
####

# Let's just call the notify() method to see all the charts in their unchanged state
m1.notify()

# Now Let's change the properties of our first model
# Change 1 triggers all 4 views and updates their labels
m1.labels = ['C#', 'PHP', 'JavaScript', 'ASP', 'Python', 'Smalltalk']
# Change 2 triggers all 4 views and updates their data
m1.data = [1, 18, 8, 60, 3, 1]
```

```python
"""
Example modified from Meier, B (2017) Python GUI Programming Cookbook, Second Edition. Packt Publishing
Refactored by Justin Cross November 2023
"""
import tkinter as tk
from tkinter import ttk


class ButtonBase(tk.Button):
    title = 'Button'
    relief = 'flat'
    foreground = 'white'

    # The constructor requires the LabelFrame object where the button will be placed together with its row position
    def __init__(self, frame, row):
        # Call the tk.Button constructor passing in the required parameters
        super().__init__(frame, text=self.title, relief=self.relief, foreground=self.foreground)
        self.grid(column=0, row=row)


# Concrete Product buttons
class ButtonRidge(ButtonBase):
    title = 'Ridge Button'
    relief = 'ridge'
    foreground = 'red'


class ButtonSunken(ButtonBase):
    title = 'Sunken Button'
    relief = 'sunken'
    foreground = 'blue'


class ButtonGroove(ButtonBase):
    title = 'Groove Button'
    relief = 'groove'
    foreground = 'green'


###### I Created NEW CLASS##########
####This class is for creating new entry widgets with default text and background, and location (frame row)

class TextBase(tk.Entry):
    textvariable = 'default text'
    background = 'default background'


    def __init__(self, frame, row):
 # Provides an Entry widget with a given LabelFrame and row position.
        tk_string = tk.StringVar()
        tk_string.set(self.textvariable)
        super().__init__(frame, textvariable=tk_string, background=self.background, foreground="white")
        self.grid(column=1, row=row)
#################################

#### CREATED NEW CLASSES ###########
# They  extend TextBase and represent concrete text widgets with specific attributes

class RedTextInput(TextBase):
    textvariable = 'red text'
    background = 'red'


class BlueTextInput(TextBase):
    textvariable = 'blue text'
    background = 'blue'


class GreenTextInput(TextBase):
    textvariable = 'green text'
    background = 'green'


##########################################


class ButtonFactory():
    _button_classes = [ButtonRidge, ButtonSunken, ButtonGroove]

    # The factory method
    def create_button(self, frame, index):
        # Get the class for the type of button we need to create
        button_class = self._button_classes[index]
        # Call the constructor of the class with the required frame and row position
        return button_class(frame, index)


############# CREATED NEW CLASS##################
##This class creates specific Entry widgets using add_new_text_entry based on index.

class TextFactory():
    _text_classes = [RedTextInput, BlueTextInput, GreenTextInput]

    # The factory method
    def add_new_text_entry(self, frame, index):
        text_class = self._text_classes[index]
        return text_class(frame, index)

##################################################


class OOP():
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Python GUI")
        self.create_widgets()

    def create_widgets(self):
        self.frame = ttk.LabelFrame(text=' Widget Factory ')
        self.frame.grid(padx=40, pady=10)
        self.create_buttons()
        self.add_new_text_entry_fields()  ##### Added new text entry method

    def create_buttons(self):
        factory = ButtonFactory()
        factory.create_button(self.frame, 0)
        factory.create_button(self.frame, 1)
        factory.create_button(self.frame, 2)

 ###### created new text entry method
    def add_new_text_entry_fields(self):  #
        factory = TextFactory()  #
        factory.add_new_text_entry(self.frame, 0)
        factory.add_new_text_entry(self.frame, 1)
        factory.add_new_text_entry(self.frame, 2)


# ==========================
oop = OOP()
oop.window.mainloop()

  
```

```python

#### PRODUCTS

class Animal: #Abstract Product
    def speak(self):
        pass
    def __str__(self):
        pass

class Dog(Animal): # Concrete Product
    def speak(self):
        return "Woof!"

    def __str__(self):
        return "Dog"

class Cat(Animal): #Concrete Product
    def speak(self):
        return "Meeoowww!"

    def __str__(self):
        return "Cat"


#### FACTORY

class AnimalFactory: # Abstract Factory

    def get_pet(self):
        pass

    def get_food(self):
        pass

class DogFactory(AnimalFactory): # Concrete Factory
    def get_pet(self):
        return Dog() # Returns dog object

    def get_food(self):
        return "Dog Food!" # Returns dog food object


class CatFactory(AnimalFactory): # Concrete Factory
    def get_pet(self):
        return Cat() #Returns cat object

    def get_food(self):
        return "Cat Food!" #Returns cat food object


### CLIENT
class PetStore: #Contains abstract method
    def __init__(self, pet_factory=None):
        self._pet_factory = pet_factory #Abstract method

    def show_pet(self): #Displays detials of pet
        pet = self._pet_factory.get_pet()
        pet_food = self._pet_factory.get_food()
        print("Our pet is '{}'!".format(pet))
        print("Our pet says hello by '{}'".format(pet.speak()))
        print("Its food is '{}'!".format(pet_food))


# Dog Demo
dog_factory = DogFactory() # Creates concrete factory
dog_shop = PetStore(dog_factory) #Creates pet store containg the Abstract Factory.
dog_shop.show_pet() # Calls utility methods to show details of dog

print("  ") #Just for readability

# Cat  Demo
cat_factory = CatFactory()
cat_shop = PetStore(cat_factory) #Creates pet store containg the Abstract Factory.
cat_shop.show_pet() # Calls utility methods to show details of cat
```

(Observer Design Pattern)

#### Intent ###

The Observer Design Pattern offers a structured solution to maintaining consistency across related objects without creating excessive coupling between them. It enables an the subject to automatically notify changes to a list of observers. The primary objective of the Observer Design Pattern is to create a one-to-many dependency between objects, where a change in the state of one object automatically notifies and updates all its observer objects.

### Motivation ###

The Observer pattern is a useful design pattern for situations where multiple objects need to be notified about changes from a particular object. For example a mobile email app notification badge. When a user receives new emails, they want to be notified of the new mail. Without the Observer pattern, the server would face challenges in creating and checking for new emails. The Observer pattern simplifies the process by easily sending notifications whenever a new email (a chnage in state)  is received.

(Taken from example on logbook slide)

### Implementation ###

For instance, you could design a system using the Observer design pattern for a music streaming app, such as Spotify’s New Music Radar Notification service. Users can follow their favorite artists and receive notifications whenever a new song is published, along with its title.

In Python, the Observer pattern can be implemented using a base Subject class that manages a list of observers (attach, detach, and notify methods). Then Concrete subjects, like Spotify, extend this base class and define state changes (e.g., new music is published). Finally Observer classes implement an update method to respond to notifications.

### Code ###

```python
# THE SUBJECT (Interface to attach, detach, and notify observer objects.)
class Subject(object):  # This represents what is being observed
    def __init__(self):
        self._observers = []  # All refrences to the observers are kept here.

    def attach(self, observer):
        if observer not in self._observers:
            self._observers.append(observer)  # Adds the observer to the list if not already in list

    def detach(self, observer):  # Removes remove the observer
        try:
            self._observers.remove(observer)
        except ValueError:
            pass

    def notify(self, modifier=None):
        for observer in self._observers:  # Goes through all the observers in the list
            if modifier != observer:  # Excludes notifying the observer that is actually updating
                observer.update(self)  # Alerts the observers


# A CONCRETE SUBJECT (A setter interface to notify registered objects of changes).
class Spotify(Subject):  # Inherits from the Subject class
    def __init__(self):
        super().__init__()
        self._new_song_title = None  # Initialises the new song title

    @property
    def new_song_title(self):  # Getter for the new song title
        return self._new_song_title

    @new_song_title.setter
    def new_song_title(self, title):  # Setter for the new song title
        self._new_song_title = title
        self.notify()  # Notifies observers when a new song title is set


# CONCRETE OBSERVER (Listens for changes using update() method)
class Follower:
    def __init__(self, name=""):
        self._name = name  # Sets name of follower

    def update(self, subject):  # Alert method that will be called when the notify() method in the subject is called
        print(f"Follower {self._name}: A new song titled '{subject.new_song_title}' has been released!")


# Creates Subject
## For simplicity sakes in this implementation there is only 1 artist.
artistSpotifyAccount = Spotify()

# Creates Observers
follower1 = Follower("Zion Tusk")
follower2 = Follower("Jill Bates")

# Attaches observers to the Spotify account
artistSpotifyAccount.attach(follower1)
artistSpotifyAccount.attach(follower2)

# Creates New Song Release
artistSpotifyAccount.new_song_title = "Quick Question"
artistSpotifyAccount.new_song_title = "Yesterday"
```

### Evaluation ###

Python’s dynamic nature and support for class functions makes using the Observer pattern easier, especially with features such as method overriding. One advantage of using python is that it provides a easy way to managing state change communication between objects. However, one downside is that if there are numerous subjects and observers involved it can become complex and harder to manage.

(Strategy Design Pattern)

### Intent ###

The Strategy pattern offers an adjustable approach to creating and selecting a group of algorithms. One advantage of this design is that it simplifies the process of swapping out different strategies without affecting the rest of the code. This pattern make it easier to manage and extened of multiple algorithms (stratergies), as changes can be made without impacting the rest of the code. 

### Motivation ###

The Strategy pattern is a valuable design pattern for situations where you have multiple ways of performing an operation and want to switch between them easily. For example, an airport shop’s payment system, needs to handle payments in various currencies, each with different conversion rates and transaction fees. Hardcoding these conversion strategies into the codebase would make time consusimng and  difficult to maintain or extend.

By using the Strategy pattern, the program can dynamically select and switch between different currency conversion strategies without modifying the main code. This makes the system flexible and adaptable to changes, such as adding new currencies or modifying conversion rates and fees.

### Implementation ###

For instance, you could design a system using the strategy design pattern for a POS terminal at WordDutyFree in Heathrow. Since many customers come from various countries, the system could use different conversion rates and transaction fees based on the input currency when making a purchase. To achieve this, the system could be implemented by defining an abstract class that specifies the conversion algorithm interface. This abstract class would then have implementations of the algorithm, each providing a specific calculation strategy for different currencies.

```python
import types #Imports the type module

# Strategy Base Class (for currency conversion)
class DutyFreeCurrencyConversionStrategy:
    def __init__(self, function=None):
        self.name = "Default Conversion Strategy"


        if function: # Will overide the convert() method with the given function when reference is present.
            self.convert = types.MethodType(function, self) # Binds the provided function as the convert method for this instance

    def convert(self, amount): # This is the conversion method - Will be replaced by specific concrete strategy implementation
        pass


# Concrete Strategy for GBP Payment
def gbp_conversion(self, amount): #As in my example GBP is default it applies this stratergy but price stays same.
    print(f"{self.name} converting {amount} to GBP")
    exampleConversionRate = 1  # As this is the default currency.
    transactionFee = 0  # No Transaction fee for GBP as is default currency.
    return (amount * exampleConversionRate) - transactionFee

# Concrete Strategy for USD Payment
def usd_conversion(self, amount): #Converts price into USD with a specific conversion rate and Transaction fee
    print(f"{self.name} converting {amount} to USD")
    exampleConversionRate = 0.75  #example exchnage rate
    transactionFee = 5  # Transaction Transaction fee for USD.
    return (amount * exampleConversionRate) - transactionFee


# Concrete Strategy for EUR Payment
def eur_conversion(self, amount):#Converts price into USD with a specific conversion rate and Transaction fee
    print(f"{self.name} converting {amount} to EUR")
    exampleConversionRate = 0.65  # example conversion rate
    transactionFee = 3  # Transaction fee for EUR
    return (amount * exampleConversionRate) - transactionFee




# Payment Processing Class
class PaymentProcessor: #Simplifed 'payment processing' for simulation purposes
                        # Using a specific conversion strategy.

    def __init__(self, dutyFreeConversionMethod):
        self.dutyFreeConversionMethod = dutyFreeConversionMethod

    def process_payment(self, amount): #Processes the payment using the specific conversion strategy
        print(f"Processing card payment for {amount} with strategy: {self.dutyFreeConversionMethod.name}")

        # Converts the amount using the strategy's conversion method
        converted_amount = self.dutyFreeConversionMethod.convert(amount)

        # Simplifed 'payment processing' for simulation purposes
        print(f"Final payment amount after conversion and transaction fees: {converted_amount}")
        print("Payment processed successfully!\n")


# Below creates different currency conversion strategies

gbp_strategy = DutyFreeCurrencyConversionStrategy(gbp_conversion)  # Creates a new GBP conversion strategy
gbp_strategy.name = "GBP Conversion Strategy"  # Sets name for new conversion strategy

eur_strategy = DutyFreeCurrencyConversionStrategy(eur_conversion)  # Create a new eur conversion strategy
eur_strategy.name = "EUR Conversion Strategy"  # Sets name for new conversion strategy

usd_strategy = DutyFreeCurrencyConversionStrategy(usd_conversion)  # Creates a new USD conversion strategy
usd_strategy.name = "USD Conversion Strategy"  # Sets name for new conversion strategy



# Creates a payment processor and process payments using different strategies
payment_processor = PaymentProcessor(gbp_strategy)  # Uses GBP strategy for processing
payment_processor.process_payment(100)  # Processes payment for an amount of 100

payment_processor.dutyFreeConversionMethod = eur_strategy  # Switches to EUR strategy
payment_processor.process_payment(100)  # Processes payment for an amount of 100

payment_processor.dutyFreeConversionMethod = usd_strategy  # Switches to USD strategy
payment_processor.process_payment(100)  # Processes payment for an amount of 100
```
